```python
import re
import os
import pandas as pd
```

```python
file_name = "DICCIONARIO-WARAO.txt"
input_path = os.path.join("input", file_name)


with open(file_name, "r", encoding="utf-8") as f:
    lines = f.readlines()
```

```python
for line in lines:
  print(line)
```

```python
spanish_words_list = []
warao_words_list = []
```

```python
for line in lines:
    # split line by two or more dots
    parts = re.split(r'\.{2,}', line.strip())

    # check there are exactly two parts after splitting
    if len(parts) == 2:
        spanish_word = parts[0].strip()
        warao_word = parts[1].strip()

        # remove non-word characters and bullet point from Spanish words
        spanish_word = re.sub(r'[^\w\s]|[\uf0b7]', '', spanish_word)
        # remove non-word characters from Warao words
        warao_word = re.sub(r'[^\w\s]', '', warao_word)

        # append to lists if both parts arent empty after cleaning
        if spanish_word and warao_word:
            spanish_words_list.append(spanish_word)
            warao_words_list.append(warao_word)

# print lists to verify
print("Spanish words list:")
print(spanish_words_list)
print("\nWarao words list:")
print(warao_words_list)
print(f"\nLength of Spanish list: {len(spanish_words_list)}")
print(f"Length of Warao list: {len(warao_words_list)}")
```

```python
# Create a DataFrame from the two lists
df = pd.DataFrame({'Spanish': spanish_words_list, 'Warao': warao_words_list})

# Export the DataFrame to a CSV file
df.to_csv('warao_dictionary.csv', index=False, encoding='utf-8')
```
