```python
spanish_words_list = []
warao_words_list = []
for line in lines:
    # Split the line by two or more dots
    parts = re.split(r'\.{2,}', line.strip())

    # Check if there are exactly two parts after splitting
    if len(parts) == 2:
        spanish_word = parts[0].strip()
        warao_word = parts[1].strip()

        spanish_words = []
        # if multiple spanish or Warao meanings split into different rows
        if '/' in spanish_word:
          spanish_words = spanish_word.split('/')
          for i in range(len(spanish_words)):
            wrd = spanish_words[i]
            if '(' in wrd:
              wrd = re.sub(r'\([^)]*\)', '', wrd)
        elif '(' in spanish_word:     # same as before but if delimeter is '()'
          spanish_words = spanish_word.split('(')
        else:
          spanish_words.append(spanish_word)

        warao_words = []
        if '/' in warao_word:
          warao_words = warao_word.split('/')
          for i in range(len(warao_words)):
            wrd = warao_words[i]
            if '(' in wrd:
              wrd = re.sub(r'\([^)]*\)', '', wrd)
        elif '(' in warao_word:
          warao_words = warao_word.split('(')
        else:
          warao_words.append(warao_word)

        # Append to the lists if both parts are not empty after cleaning
        for spanish_word in spanish_words:
          for warao_word in warao_words:
            # Remove the bullet point from Spanish words
            spanish_word = re.sub(r'[\uF0B7()]', '', spanish_word)
            # # Remove non-word characters from Warao words
            warao_word = re.sub(r'[()]', '', warao_word)
            if spanish_word and warao_word:
              spanish_words_list.append(spanish_word.strip())
              warao_words_list.append(warao_word.strip())

# Print the lists to verify
print("Spanish words list:")
print(spanish_words_list)
print("\nWarao words list:")
print(warao_words_list)
print(f"\nLength of Spanish list: {len(spanish_words_list)}")
print(f"Length of Warao list: {len(warao_words_list)}")
```

```python
import pandas as pd

# Create a DataFrame from the two lists
df = pd.DataFrame({'Warao': warao_words_list, 'Spanish': spanish_words_list})

display(df.head(20))
```

```python
# Export the DataFrame to a CSV file
df.to_csv('./output/warao_dictionary_final.csv', index=False, encoding='utf-8')
```
