```python
# ! pip install pdfplumber
import pdfplumber
import os 
```

```python
def extract_columns(pdf_path):
    text_output = ""
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages, start=1):
            width = page.width
            height = page.height

            # Define bounding boxes (left, top, right, bottom)
            left_bbox = (0, 0, width / 2, height)
            right_bbox = (width / 2, 0, width, height)

            # Extract each column separately
            left_text = page.within_bbox(left_bbox).extract_text(x_tolerance=2, y_tolerance=2)
            right_text = page.within_bbox(right_bbox).extract_text(x_tolerance=2, y_tolerance=2)

            # Combine columns in desired reading order
            combined = ""
            if left_text:
                combined += left_text.strip() + "\n\n"
            if right_text:
                combined += right_text.strip() + "\n\n"

            text_output += f"\n{combined}\n"

    return text_output
```

```python
# Example use
cwd = os.path.abspath(os.getcwd())
os.chdir(cwd)  # changes cwd to script's directory
data_dir = "data"
input_filename = "waraotumaadeniabu-completo_historias_warao.pdf"
input_doc_path = os.path.join(cwd, data_dir, input_filename)
pdf_text = extract_columns(input_doc_path)
```

```python
# save result to file
output_filename = "processed_stories.txt"
output_path = os.path.join(data_dir, output_filename)
with open(output_path, "w", encoding="utf-8") as f:
    f.write(pdf_text)
```
