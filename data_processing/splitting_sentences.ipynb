# Sentence Splitter
Simple script to split paragraphs of sentences into individual sentences

```python
import pandas as pd
import os
import re
from transformers import pipeline
from langdetect import detect
```

```python
cwd = os.path.abspath(os.getcwd())
os.chdir(cwd)  # changes cwd to script's directory
data_dirname = "data"
data_dir = os.path.join(cwd, data_dirname)
# create data directory if it doesn't exist
if not os.path.exists(data_dir):
    os.makedirs(data_dir)
input_filename = "processed_stories.txt"
processed_stories_path = os.path.join(data_dir, input_filename)
with open(processed_stories_path, "r", encoding="utf-8") as f:
    text = f.read()
    start_match = re.search(r"dedicatoria", text, flags=re.IGNORECASE)
    end_match = re.search(r"Ã­ndice", text, flags=re.IGNORECASE)
    if start_match and end_match:
        text = text[start_match.end():end_match.start()]
    
    text = text.replace("-\n", "")   # remove word splits across lines
    text = text.replace("\n", " ")   # replace newlines with spaces

    sentences = text.lower().strip().split(". ")  # simple sentence splitting by period followed by space

print(sentences)  # Display the first few sentences
```

```python
warao_sentences = []
spanish_sentences = []
for s in sentences:
    language = detect(s)
    if language == "es":
        spanish_sentences.append(s)
    else: 
        warao_sentences.append(s)

print(f"Spanish sentences: {spanish_sentences} \n Warao sentences: {warao_sentences}" )
df = pd.DataFrame({
    "warao_sentences": warao_sentences,
    "spanish_sentences": spanish_sentences
})
df.head(10)
```

```python
df.to_csv(os.path.join(data_dir, "warao_spanish_stories_sentences.csv"), index=False)
```
